# Data Visualization

```python
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
import plotly.express as px
import cufflinks as cf
import plotly
import plotly.graph_objects as go
import plotly.offline as pyo
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.subplots import make_subplots
init_notebook_mode(connected=True)
cf.go_offline()
```

## List of Orders Exploration

```python
df_list_order = pd.read_csv('List of Orders.csv')
df_list_order.head()
```

```python
df_list_order.info()
```

```python
df_list_order = df_list_order.dropna(axis=0)
```

```python
df_list_order['Order Date'] = pd.to_datetime(df_list_order['Order Date'], format='%d-%m-%Y')
```

```python
df_list_order.info()
```

```python
df_list_order.head()
```

## Line plot : Number of order from each city

```python
sr_city = df_list_order.groupby('City').size().reset_index()
sr_city.columns = ['City','Count']

px.bar(sr_city, x='City', y='Count', color='City', title='Order count form each city', 
       labels={ 'City':'City', 'Count':'Number of Orders' })
```

### Line Plot: Order in 2019

```python
df_order_list_2019 = df_list_order[ (df_list_order['Order Date'] >= '01-01-2019') & (df_list_order['Order Date'] <= '05-01-2019') ]
sr_order_count_date = df_order_list_2019.groupby('Order Date').size().reset_index()
sr_order_count_date.columns = ['Order Date', 'Count']
```

```python
px.line(sr_order_count_date, x='Order Date', y='Count', labels={ 'x' : 'Order Date', 'y': 'Count'}, 
           title = 'Order count in 2019')
fig= go.Figure()
fig.add_trace(go.Scatter(x=sr_order_count_date['Order Date'], y = sr_order_count_date['Count'], mode='lines+markers'))
```

## Data 2 : Housing Price Data

```python
df_house_price = pd.read_csv('HousePriceData.csv')
df_house_price.head()
```

```python
df_house_price.info()
```

```python
col_selected = ['LotArea',
 'OverallQual',
 'YearBuilt',
 'YearRemodAdd',
 'MasVnrArea',
 'BsmtFinSF1',
 'BsmtUnfSF',
 'TotalBsmtSF',
 '2ndFlrSF',
 'GrLivArea']  
df_house_price_sel = df_house_price[col_selected]
df_house_price_sel['SalePrice'] = df_house_price['SalePrice']
```

```python
df_house_price_sel = df_house_price_sel.dropna(axis=0)
df_house_price_sel.info()
```

### Scatter plots

```python
fig = make_subplots(rows=4, cols=3, subplot_titles=[str(col+ ' vs Sales Price') for col in col_selected], 
                     x_title = 'Sales Price', y_title = 'Features')
col_num = 0
row_num = 1
for col in col_selected:
    if col_num == 3:
        col_num = 0
        row_num = row_num +1
    col_num += 1 
    title = col + ' V/s ' + 'Sales Price'
    fig.add_trace(go.Scatter(x=df_house_price_sel.SalePrice, y=df_house_price_sel[col], mode='markers', name=title), 
                  row=row_num, col=col_num, 
                   )
#    fig.update_yaxes(title_text=col, row=row_num, col=col_num)

fig.update_layout(height=1000, width=1000, title_text="Sales Price vs all Features",
                  annotations=[{ 
                      'font': {'size': 10}
                  }])

fig.show()    
```
